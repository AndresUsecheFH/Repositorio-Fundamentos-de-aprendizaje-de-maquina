Eliminacion de valores nulos en columnas y llenado de columnas con valores especificos

Cabe resaltar que, dado que no se sabe para que se quiere analizar la base, no se sabe que variables son mas importantes que otras, ni que porcentaje de datos nulos se puede tolerar, por lo cual no eliminamos muchas vaariables y se rellenan algunos valores nulos con "nothing description"

```python
#eliminacion para solo columnas

precio_casa.drop(["Dimensions", "Plot Area", ], axis=1, inplace=True)    #axis = columnas o filas -> inplace = modificar el dataset original
```

```python
#Algunas columnas numéricas están en formato object en lugar de int o float, lo que puede causar errores.

precio_casa["Bathroom"] = pd.to_numeric(precio_casa["Bathroom"], errors="coerce")
precio_casa["Balcony"] = pd.to_numeric(precio_casa["Balcony"], errors="coerce")

```

```python
#llenar los espacios vacios con algun valor en especificio

precio_casa["Price (in rupees)"] = precio_casa["Price (in rupees)"].fillna(0) #en este caso con 0
precio_casa["Bathroom"] = precio_casa["Bathroom"].fillna('NA') #en este caso con NA
precio_casa["Description"] = precio_casa["Description"].fillna('nothing description') #en este caso con strings
precio_casa["Car Parking"] = precio_casa["Car Parking"].fillna('nothing description')
precio_casa["Super Area"] = precio_casa["Super Area"].fillna('nothing description')
#y lo mismos con las demas columnas
```

```python
# observar nuevamente los resultados

print(precio_casa.isnull().sum())
```

Normalizacion y/o estandarizacion

```python
precio_casa.head()
```

Separamos las unidades de medida de los valores de Tile, Amount(in rupees) y Carpet Area

```python
df = pd.DataFrame(precio_casa)
df[['Amount', 'Rupees']] = df['Amount(in rupees)'].str.extract(r'(\d*\.?\d+)\s*(\D+)')
df[['BHK', 'Status', 'Type', 'Location']] = df['Title'].str.extract(r'(\d+\s*BHK)\s*(.*?)\s*(Flat|Apartment|Villa|House)\s*(?:for\s*\w*\s*in\s*(.*))?')

# Convertir BHK, Amount y Carpert area a numero
df['BHK'] = df['BHK'].str.extract(r'(\d+)').astype(float)
df['Amount'] = pd.to_numeric(df['Amount'])
df ['Carpert area'] = pd.to_numeric(df['Carpet area'])

df[['Carpet area', 'unit of measurement']] = df['Carpet Area'].str.extract(r'(\d*\.?\d+)\s*(\D+)')

df.head()
```

```python
#Se eliminan las columnas
df.drop(["Amount(in rupees)" ], axis=1, inplace=True)
df.drop(["Carpet Area"], axis=1, inplace=True)
```

Normalización, División en conjuntos de entrenamiento y prueba

```python
from sklearn.preprocessing import StandardScaler

# Definir las columnas numéricas que quedan después de la limpieza
numeric_columns = ["Price (in rupees)", "Carpet area"]

# Inicializar el escalador
scaler = StandardScaler()

# Aplicar estandarización solo a las columnas numéricas restantes
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])

# Mostrar los primeros resultados
print(df[numeric_columns].head())




```

```python
from sklearn.model_selection import train_test_split

# Separar las características (X) y el objetivo (y)
X = df.drop(["Price (in rupees)"], axis=1)  # Aquí elimina la columna objetivo
y = df["Price (in rupees)"]  # Esta es la variable objetivo

# Dividir los datos en conjuntos de entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Mostrar las dimensiones de los conjuntos resultantes
print(f"Conjunto de entrenamiento: {X_train.shape}, {y_train.shape}")
print(f"Conjunto de prueba: {X_test.shape}, {y_test.shape}")
```
